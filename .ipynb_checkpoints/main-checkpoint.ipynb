Galiausiai, paskutinis ≈°ios analizƒós tikslas yra sukurti modelƒØ gebantƒØ nustatyti i≈° turim≈≥ komandos statistikos ar komanda laimƒójo rungtynes ar ne. 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer
from concurrent.futures import ThreadPoolExecutor
import time
import tensorflow as tf

team_stats = games_df.groupby(['GAME_ID', 'TEAM']).agg({
    'FG': 'sum',
    'FGA': 'sum',
    'FG3': 'sum',
    'FG3A': 'sum',
    'FT': 'sum',
    'FTA': 'sum',
    'ORB': 'sum',
    'DRB': 'sum',
    'TRB': 'sum',
    'AST': 'sum',
    'STL': 'sum',
    'BLK': 'sum',
    'TOV': 'sum',
    'PF': 'sum',
    'PLUS_MINUS': 'sum',
    'PTS': 'sum',
    'TOTAL_MINS': 'first',
    'RESULT': 'first',
}).reset_index()

# Apskaiƒçiuojame procentines komandos vertes
percentage_vars = ['FG%', 'FG3%', 'FT%']
for var in percentage_vars:
    team_stats[var] = team_stats[f'{var[:-1]}'] / team_stats[f'{var[:-1]}A']

print("++++++++ Komandos statistika (pirmu 3) ++++++++")
print(team_stats.head(3))
# Spƒójimas atliekamas ar RESULT bus W (pergalƒó) ar L(pralaimƒójimas)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(team_stats['RESULT'])

# Duomenis daliname ƒØ apmokymo ir testavimo dalis
X_team = team_stats.drop(['RESULT', 'TOTAL_MINS'], axis=1)
y_team = team_stats['RESULT'] 

numeric_columns = team_stats.select_dtypes(include=['number']).columns
X = team_stats[numeric_columns]

print(X.head(10))
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.4, random_state=42)

# Pakeiƒçiame tr≈´kstamas stulpeli≈≥ vertes su atitinkamo vertƒós stulpelio vidurkiu
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Z-score normalizacija
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# Sukuriame neuronini≈≥ tinkl≈≥ modelƒØ remiantis TensorFlow API
# Pirmas sluoksnis su 64 neuronais 
# Antras sluoksnis su 1 neuronu
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])

# Modeli≈≥ apmokymo funkcijos
def tensorflow_train_task():
    """
    Trains a TensorFlow neural network model on the scaled training data.

    Returns:
        tf.keras.Model: The trained TensorFlow neural network model.
    """
    print("Starting tensorflow_train_task")
    start_time = time.time()

    model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, verbose=0)

    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Finished tensorflow_train_task. Elapsed time: {elapsed_time:.2f} seconds")
    return model

def rf_classification_task():
    """
    Trains a Random Forest classifier on the scaled training data.

    Returns:
        RandomForestClassifier: The trained Random Forest classifier.
    """
    print("Starting rf_classification_task")
    start_time = time.time()

    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train_scaled, y_train)

    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Finished rf_classification_task. Elapsed time: {elapsed_time:.2f} seconds")
    return clf

# Asynchroni≈°kai apmokame savo TensorFlow neuroninio tinklo ir atsitiktini≈≥ mi≈°k≈≥ modelius.
with tf.device('/GPU:0'):
    with ThreadPoolExecutor() as executor:
        future_tensorflow = executor.submit(tensorflow_train_task)
        future_rf = executor.submit(rf_classification_task)

        results = [future_tensorflow.result(), future_rf.result()]

tensorflow_model, rf_model = results

# ƒÆvertiname sukurtus modelius su testavimo aibƒómis
y_pred_tensorflow = (tensorflow_model.predict(X_test_scaled) > 0.5).astype(int).flatten()
y_pred_rf = rf_model.predict(X_test_scaled)

accuracy_tensorflow = accuracy_score(y_test, y_pred_tensorflow)
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(f'TensorFlow Model Accuracy: {accuracy_tensorflow}')
print(f'Random Forest Model Accuracy: {accuracy_rf}')
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Sumai≈°imo matrica TensorFlow modeliui
cm_tensorflow = confusion_matrix(y_test, y_pred_tensorflow)
disp_tensorflow = ConfusionMatrixDisplay(confusion_matrix=cm_tensorflow)
disp_tensorflow.plot()

# Sumai≈°imo matrica atsitiktinio mi≈°ko modeliui
cm_rf = confusion_matrix(y_test, y_pred_rf)
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf)
disp_rf.plot()
```
